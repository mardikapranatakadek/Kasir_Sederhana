```python
id = ["admin"]
pas = ["admin123"]

print("""
      Selamat Datang pada tampilan home \n
      Pilih salah satu untuk melanjutkan \n
      1. Login Kasir
      2. Daftar Akun
       """)
def baru():
  id_baru = input("Masukkan Id Karyawan : ")
  pas_baru = input("Masukkan Password : ")
  id.append(id_baru)
  pas.append(pas_baru)
  print("Silakan Login Kasir \n")

tampil = True
while tampil:
  tampil1 = int(input("Masukkan Pilihan Home: "))
  if tampil1 == 1:
    tampil = False
  elif tampil1 == 2:
    baru()

  
login_try = True
while login_try:
    id_try= input("Id Karyawan : ")
    pas_try = input("Password  : ")
    if (id_try in id ) and (pas_try in pas):
        login_try = False
    else:
        login_try = True
        print("Username atau password salah!! Silahkan coba lagi")
    
def menu():
    print("=========== Warung Makan Makmur ===========")
    print("===========    DAFTAR MENU     ============")
    print("""
          MAKANAN
          1. Bakso                 10000
          2. Soto                  12000
          3. Mie Ayam              15000
          """)
    print("--------------------------------")
    print("""
          MINUMAN
          1. Es teh                3000
          2. Es Jeruk              5000
          3. Rasa-rasa             2000
          """)
menu()
run = True
while run:

    def funmakanan():
        global total1
        global porsi
        global mkn
        makanan = int(input("Masukkan Pilihan Makanan: "))
        porsi = int(input("Berapa Porsi? "))

        if makanan == 1:
            total1 = porsi * 10000
            mkn = ("Bakso")
        elif makanan == 2:
            total1 = porsi * 12000
            mkn = ("Soto")
        elif makanan == 3:
            total1 = porsi * 15000
            mkn = ("Mie Ayam")
        else:
            print("Menu kamu tidak ada didaftar")
            funmakanan()
    
    def funminuman():
        global total2
        global gelas
        global minu
        minuman = int(input("Masukkan Pilihan Minuman: "))
        gelas = int(input("Berapa Gelas? "))

        if minuman == 1:
            total2 = gelas * 3000
            minu = ("Es Teh")
        elif minuman == 2:
            total2 = gelas * 5000
            minu = ("Es Jeruk")
        elif minuman == 3:
            total2 = gelas * 2000
            minu = ("Rasa-rasa")
        else:
            print("Menu kamu tidak ada didaftar")
            funminuman()
    funmakanan()
    funminuman()
    totalbelanja = total1 + total2

    print(porsi, "Porsi", mkn, "Seharga Rp", total1)
    print(gelas, "Gelas", minu, "Seharga Rp", total2)
    print("\nTotal yang harus dibayar: Rp", totalbelanja)
    uang = int(input("Uang Tunai: Rp"))
    kembalian = uang - totalbelanja
      
    print("Kembalian : Rp", kembalian)
    print("=====================================================")
    print("================= STRUK BELANJA =====================")
    print("============== Warung Makan Makmur ==================")
    print("Belanja :\n", porsi, mkn, "(Rp", total1,           ")")
    print("\n\b", gelas, minu, "(Rp", total2,               ")\n")
    print("Tagihan \t    : Rp", totalbelanja                     )
    print("Dibayar \t    : Rp", uang                             )
    print("Kenbalian \t  : Rp", kembalian                        )
    print("=====================================================")
    print("                     Terimakasih                     ")
    print("=====================================================")

    tambahan = str(input("Apakah belanja lagi? "))

    if tambahan == "ya":
      pass
    elif tambahan == "tidak" :
      break
```
